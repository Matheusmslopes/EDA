## Transformação de Dados

```python
# Transformar coluna TotalCharges de String para Float - Abordagem 1 com astype
# df_contracts.TotalCharges (não funciona com nomes que tenham espaco) ou df_contracts['TotalCharges']
df_contracts['TotalCharges'] = df_contracts.TotalCharges.astype(float)
# Nesse caso vai dar erro pois tem espaço em branco nos registros (não consegue transformar espaço em branco em float)
```

```python
# Transformar coluna TotalCharges de String para Float - Abordagem 2 com to_numeric
df_contracts['TotalCharges'] = pd.to_numeric(df_contracts.TotalCharges)
# Nesse caso vai dar erro pois tem espaço em branco nos registros (não consegue transformar espaço em branco em float)
```

```python
# Transformar coluna TotalCharges de String para Float - Abordagem 3 com to_numerico com coerce
df_contracts['TotalCharges'] = pd.to_numeric(df_contracts.TotalCharges, errors='coerce')
# Nesse caso a transformação vai acontecer, pois onde tem espaço em branco ou algo que ele não consiga converter, vai se transformar em nulo/null
```

```python
# Mostrar estrutura / Schema do DataFrame para verificar transformação
df_contracts.info()
```

Após a transformação da coluna TotalCharges do DataFrame Contracts, a coluna passou a ter 11 valores ausentes (missing values)
