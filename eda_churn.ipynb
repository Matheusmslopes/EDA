## Análise Bivariada

#### Tabela de contingência


```python
# Hipótese: Clientes com contrato do tipo Mensal são mais propensos ao churn
# Construir tabela de contingência entre tipo de contrato e churn - Quantidade
pd.crosstab(df_churn.Churn, df_churn.Contract, margins=True, margins_name='Total')
```

```python
# Construir tabela de contingência entre tipo de contrato e churn - Proporção
pd.crosstab(df_churn.Churn, df_churn.Contract, normalize= 'index',margins=True, margins_name='Total')
```

Considerando o público total, os clientes de contrato mensal representam 55%. Porém, se considerarmos apenas os clientes que abandonaram o serviço, 88% tinha contrato mensal. Desta forma podemos afirmar que há uma correlação entre o tipo de contrato mensal e o abandono.

#### Correlação de variáveis qualitativas

```python
# Avaliar a correlação entre duas variáveis categóricas (qualitativas)
# Executar um teste de hipótese (Chi-Square/Qui-Quadrado) de Pearson
# Em um teste de hipótese, duas hipóteses são formuladas:
# H0 (Hipótese Nula): as duas variáveis são independentes
# H1 (Hipótese complementar): as duas variáveis não são independentes

# O teste serve para confirmar ou recusar a hipótese nula (H0)
# Quando a probabilidade de observarmos H0 é inferior a 0.05 (p-value), recusamos a H0 e seguimos com a H1
```

```python
# Gerar um DataFrame da Crosstab (sem totais)
df_crosstab_churn_contract = pd.crosstab(df_churn.Churn, df_churn.Contract) 
```

```python
# Calcular os Scores e P-values
from scipy.stats import chi2_contingency
```

```python
chi_scores_churn_contract = chi2_contingency(df_crosstab_churn_contract)
print(chi_scores_churn_contract)
```

```python
scores_churn_contract = pd.Series(chi_scores_churn_contract[0])
print(scores_churn_contract)
```

```python
pvalues_churn_contract = pd.Series(chi_scores_churn_contract[1])
print(pvalues_churn_contract)
```

```python
# Apresentar números com decimais sem a notação científica
pd.set_option('display.float_format', lambda x: '%.15f' % x)
```

```python
# Criar DataFrame com os resultados
df_chi_scores_churn_contract = pd.DataFrame({'Qui2': scores_churn_contract, 'p-value': pvalues_churn_contract})
```

```python
# Analisar Scores e p-values
df_chi_scores_churn_contract
```

Conforme Teste Estatístico (Chi-Square), o P-Value <= 0.05, desta forma rejeitamos a hipótese nula, ou seja, as variáveis não são independentes. Pelo Qui2 alto, podemos afirmar que há uma forte correlação.

#### Correlação de variável qualitativa e quantitativa


```python
# Avaliar a correlação entre uma variável qualitativa e quantitativa
# Hipótese: Cliente com menos de 6 meses de contrato é mais propenso ao Churn

# Criar um coluna nova no DataFrame com base em uma condição de outra coluna (TempoMenor6Meses)
import numpy as np
```

```python
# Criar coluna nova
df_churn['TimeLess6months'] = np.where(df_churn.tenure < 6, 'Yes', 'No')
```

```python
# Verificar o DataFrame
df_churn.head(5)
```

```python
# Gerar um DataFrame da Crosstab (sem totais)
df_crosstab_churn_tenure6 = pd.crosstab(df_churn.Churn, df_churn.TimeLess6months) 
```

```python
# Calcular os Scores
chi_scores_churn_tenure6 = chi2_contingency(df_crosstab_churn_tenure6)
print(chi_scores_churn_tenure6) 
```

```python
scores_churn_tenure6 = pd.Series(chi_scores_churn_tenure6[0])
print(scores_churn_tenure6)
```

```python
pvalues_churn_tenure6 = pd.Series(chi_scores_churn_tenure6[1])
print(pvalues_churn_tenure6)
```

```python
# Criar DataFrame com os resultados
df_chi_scores_churn_tenure6 = pd.DataFrame({'Qui2': scores_churn_tenure6, 'p-value': pvalues_churn_tenure6})
```

```python
# Analisar o DataFrame
df_chi_scores_churn_tenure6
```

Como o p-value <=0.05, rejeitamos a H0, as variáveis não são independentes. Vale mencionar que com base do Qui2 resultante das 2 análises, a correlação entre Churn e Tempo de contrato < 6 meses é menor do que a correlação entre Churn e Tipo de contrato

#### Método de Tukey

```python
# IQR - Range interquartil (Distribuição não normal)
# IQR = 3°quartil - 1°quartil
q1_TotalCharges_month = df_churn_month.TotalCharges.quantile(0.25)
q3_TotalCharges_month = df_churn_month.TotalCharges.quantile(0.75)
IQR_TotalCharges_month = q3_TotalCharges_month - q1_TotalCharges_month
print(IQR_TotalCharges_month)
```

```python
# Limites Inferior e Superior (Valores absolutos)
limInf_TotalCharges_month = q1_TotalCharges_month - (IQR_TotalCharges_month * 1.5)
limSup_TotalCharges_month = q3_TotalCharges_month + (IQR_TotalCharges_month * 1.5)
print(limInf_TotalCharges_month)
print(limSup_TotalCharges_month)
```

```python
df_churn_month[(df_churn_month.TotalCharges < limInf_TotalCharges_month) | (df_churn_month.TotalCharges > limSup_TotalCharges_month)]
```

#### Método Zscore

```python
# Usado para distribuição normal
# Indica quantos desvios padrões de um ponto específico de dados está distante da média
# Zscore = (x - média ) / desvio padrão
from scipy.stats import zscore
```

```python
z = np.abs(zscore(df_churn_month.TotalCharges))
```
